### Предобработка данных

#### Удаление пропусков

```python
pd.DataFrame(round(data.isna().mean()*100,)).style.background_gradient('coolwarm')
```

```python
# В столбце total_area пропусков нет. Попробуем расчитать примерные
# значения living_area и kitchen_area, посчитав их среднее соотношение к total_area.
mean_living_area = (data['living_area'] / data['total_area']).mean()
mean_kitchen_area = (data['kitchen_area'] / data['total_area']).mean()

# Заполняем пропуски
data['balcony'] = data['balcony'].fillna(0)
data['is_apartment'] = data['is_apartment'].fillna(False)
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data['living_area'] = data['living_area'].fillna(data['total_area'] * mean_living_area)
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] * mean_kitchen_area)

# for locality_name in data['locality_name'].dropna().unique():
#     median_ceiling = data.loc[data['locality_name'] == locality_name, 'ceiling_height'].median()
#     data.loc[(data['ceiling_height'].isna()) & (data['locality_name']==locality_name), 'ceiling_height'] = median_ceiling

data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())

pd.DataFrame(round(data.isna().mean()*100,)).style.background_gradient('coolwarm')
```

**Обоснования замены даных.**
- "balcony" - если число балконов не указано, то, скорее всего, в квартире их нет
- "ceiling_height" - можно заменить пропуски на медиану, на которую не влияют выбросы, которые еще не обработаны
- "is_apartment" - пропуск, скорее всего, говорит о том, что это обычная квартира. Можно заменить пропуски на False  
- "parks_around3000" - если число парков не рассчитано, то, скорее всего, их поблизости нет
- "ponds_around3000" - если число прудов не рассчитано, то, скорее всего, их поблизости нет
- "living_area" - можно заменить пропуски, рассчитав среднее соотношение living_area к total_area по всем данным.
- "kitchen_area" - можно заменить пропуски, рассчитав среднее соотношение kitchen_area к total_area по всем данным.

**Причины, которые могли привести к пропускам в данных.**
- 'balcony' - если в квартире нет балконов, то продавец, видимо, поленился указать 0, а система в этом стобце не обрабатывает пропуски
- "ceiling_height" - продавец мог не знать высоту потолка или поленился/забыл указать
- "is_apartment" - возможность указать этот параметр, вероятно, возникла спустя какое-то время после начала сбора данных. Продавцы также могли просто оставить данный параметр пустым.
- "parks_around3000" - система не смогла найти ближайший парк
- "ponds_around3000" - система не смогла найти ближайший пруд
- "living_area" - продавец забыл указать или не знал, какова жилая площадь
- "kitchen_area" - продавец забыл указать или не знал, какова площадь кухни

### Посчитать и добавить в таблицу новые столбцы

Нужно произвести следующие подсчеты:
1. цена одного квадратного метра;
2. день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
3. месяц публикации объявления;
4. год публикации объявления;
5. тип этажа квартиры (значения — «первый», «последний», «другой»);
6. расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).

```python
# цена одного квадратного метра
data['square_meter_price'] = data['last_price'] / data['total_area']

# день недели публикации объявления
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday

# месяц публикации объявления
data['month_exposition'] = data['first_day_exposition'].dt.month

# год публикации объявления
data['year_exposition'] = data['first_day_exposition'].dt.year

def floor_category(row):
    total_floors = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == total_floors:
        return 'последний'
    elif 1 < floor < total_floors:
        return 'другой'

# тип этажа квартиры  
data['floor_category'] = data.apply(floor_category, axis=1)

# расстояние до центра города в километрах
data['cityCenters_nearest_km'] = (data['cityCenters_nearest'] / 1000).round(0)
```

#### Какие факторы больше всего влияют на общую (полную) стоимость объекта?

Постройте графики, которые покажут зависимость цены от указанных выше параметров. Для подготовки данных перед визуализацией вы можете использовать сводные таблицы.

```python
corr_data = ['total_area', 'living_area', 'kitchen_area', 'rooms',
            'weekday_exposition', 'month_exposition', 'year_exposition']

for i in corr_data:
    pt = data.pivot_table(index = i, values = 'last_price', aggfunc = ['median'])
    pt.columns = ['median']
    pt.plot(y = 'median', style = 'o', figsize=(4,4), title=f'корреляция в {i} равна {data[i].corr(data["last_price"]).round(2)}')
```

```python
data.pivot_table(index='floor_category', values='last_price', aggfunc='median').plot(y='last_price');
```

**Вывод:**  
Самая большая корреляция с общей стоимостью жилья у столбцов `total_area`, `living_area`, `kitchen_area` - она составляет 0.65, 0.58 и 0.52 соответственно. Также есть корреляция между ценой и этажом - заметно снижение цены у квартир на первом и последнем этаже.

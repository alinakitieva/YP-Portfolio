**Тест Дики-Фуллера**

Нулевая гипотеза - Временной ряд не стационарен

Альтернативная гипотеза - Временной ряд стационарен

```python
result = adfuller(data['num_orders'])

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
alpha = 5
if result[0] < result[4][str(alpha)+"%"]:
    print ("Нулевая гипотеза отвергнута – Временной ряд стационарен")
else:
    print ("Нулевая гипотеза не отвергнута – Временной ряд не стационарен")
```

```python
alpha = 1
if result[0] < result[4][str(alpha)+"%"]:
    print ("Нулевая гипотеза отвергнута – Временной ряд стационарен")
else:
    print ("Нулевая гипотеза не отвергнута – Временной ряд не стационарен")
```

В ходе анализа были выполнены два теста стационарности. Результаты теста Дики-Фуллера и анализа скользящего среднего могут быть сравнены следующим образом:

- Скользящее среднее: сглаженный ряд показывает тренд, и это указывает на нестационарность исходного временного ряда, так как среднее значение ряда меняется со временем.

- Тест Дики-Фуллера: при уровне значимости, равном 5%, нулевую гипотезу о нестационарности ряда можно отвергнуть, и, следовательно, временной ряд является стационарным. При уровне значимости, равном 1%, нулевую гипотезу о нестационарности ряда  отвергнуть не получается, и, следовательно, временной ряд не является стационарным. 

С точки зрения анализа скользящего среднего временной ряд нестационарный. С точки зрения теста Дики-Фуллера временной ряд может быть как стационарным, так и нестационарным в зависимости от от уровня значимости.

## Обучение

### Создание признаков

```python
def make_features(df, max_lag, rolling_mean_size):
    features_df = df.copy()
    features_df['day'] = features_df.index.day
    features_df['hour'] = features_df.index.hour
    features_df['dayofweek'] = features_df.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        features_df['lag_{}'.format(lag)] = features_df['num_orders'].shift(lag)

    features_df['rolling_mean'] = features_df['num_orders'].shift().rolling(
        rolling_mean_size).mean()
    
    return features_df

data = make_features(data, 7, 7)
data.head(5)
```

### Разделение на выборки

```python
train, test = train_test_split(data, shuffle=False, test_size=0.1)
train = train.dropna()

X_train, Y_train = train.drop('num_orders', axis=1), train['num_orders']
X_test, Y_test = test.drop('num_orders', axis=1), test['num_orders']
```

### Стандартизация

```python
# Создаем объект стандартизатора
scaler = StandardScaler()

# Обучаем стандартизатор на обучающем наборе и применяем его к обучающему и валидационному набору
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Линейная регрессия

```python
tscv = TimeSeriesSplit(n_splits=3)
```

```python
linear_model = LinearRegression()

# Проведение кросс-валидации с 3 фолдами
cross_val_scores = cross_val_score(linear_model, X_train, Y_train,
                                   cv=tscv, scoring='neg_root_mean_squared_error')

cross_val_rmse = -cross_val_scores

print("Cross-validation RMSE for Linear Regression: ", cross_val_rmse)

# Вывод среднего значения и стандартного отклонения RMSE по всем фолдам
print("Mean cross-validation RMSE: ", cross_val_rmse.mean())
print("Standard Deviation of cross-validation RMSE: ", cross_val_rmse.std())
```

### LightGBM

```python
data_lgbm = data.astype({"hour":'category',
                         "day":'category',
                         "dayofweek":'category'})
```

```python
(X_train_lgbm, X_test_lgbm,
 Y_train_lgbm, Y_test_lgbm) = train_test_split(data_lgbm.drop("num_orders", axis = 1),
                                               data_lgbm["num_orders"],
                                               test_size = 0.1,
                                               random_state = 12345)
```

```python
# Параметры для поиска
param_grid = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1],
    'max_depth': [-1, 10],
}

lgbm = LGBMRegressor(random_state=12345)

grid_search = GridSearchCV(estimator=lgbm, param_grid=param_grid,
                           cv=tscv, scoring='neg_root_mean_squared_error', n_jobs=-1)

# Обучаем модель
grid_search.fit(X_train_lgbm, Y_train_lgbm)

# лучшие гиперпараметры
best_parameters = grid_search.best_params_
print("Best parameters: ", best_parameters)
# Вывод наилучшего RMSE
print('Best RMSE:', -grid_search.best_score_)
```

### DecisionTreeRegressor

```python
model = DecisionTreeRegressor(random_state=12345)

# Определение параметров для GridSearchCV
params = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_leaf': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 10, 15]
}

# Используем RMSE как метрику для оптимизации
rmse_scorer = make_scorer(lambda y_true, y_pred: mean_squared_error(y_true, y_pred)**0.5,
                          greater_is_better=False)

# Определение и обучение GridSearchCV
grid = GridSearchCV(model, params, cv=tscv, scoring=rmse_scorer, n_jobs=-1)
grid.fit(X_train, Y_train)

# Вывод наилучших параметров
print('Best parameters:', grid.best_params_)

# Вывод наилучшего RMSE
print('Best RMSE:', -grid.best_score_)
```

### Вывод

Мы обучили и оценили качество трех различных моделей: линейной регрессии, градиентного бустинга LightGBM и решающего дерева. Все модели прошли процесс кросс-валидации, после которого было вычислено среднее значение RMSE.

Линейная регрессия: Среднее значение RMSE на кросс-валидации составило около 33.95, что ниже требуемого значения метрики (не больше 48).

LightGBM: При оптимальных параметрах обучения (learning_rate=0.1, max_depth=10, n_estimators=50) модель показала RMSE равный 25.5 на кросс-валидации. Это гораздо меньше, чем у линейной регрессии, и соответствует условию задачи.

Решающее дерево: При оптимальных параметрах (max_depth=15, min_samples_leaf=20, min_samples_split=2) RMSE на кросс-валидации составило 29.39. Хотя это больше, чем у модели LightGBM, оно все еще удовлетворяет условию задачи.

Исходя из полученных результатов, наиболее эффективной моделью оказалась модель градиентного бустинга LightGBM. Она показала наименьшее значение RMSE.

## Тестирование

### LightGBM

```python
# Обучаем модель с использованием лучших гиперпараметров
model_lgbm_best = LGBMRegressor(n_estimators=best_parameters['n_estimators'],
                                learning_rate=best_parameters['learning_rate'],
                                max_depth=best_parameters['max_depth'],
                                loss_function='RMSE', 
                                random_seed=12345)

# Обучение модели
model_lgbm_best.fit(X_train_lgbm, Y_train_lgbm)

# Предсказания на тестовой выборке
predictions_test = model_lgbm_best.predict(X_test_lgbm)

# RMSE на тестовой выборке
rmse_test = np.sqrt(mean_squared_error(Y_test_lgbm, predictions_test))

print("RMSE on test data:", rmse_test)
```

### Проверка на адекватность с помощью DummyRegressor

```python
# Создаем и обучаем DummyRegressor
dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(X_train, Y_train)

# Предсказываем значения на тестовой выборке
dummy_predictions = dummy_regr.predict(X_test)

# Вычисляем RMSE для DummyRegressor
dummy_rmse = np.sqrt(mean_squared_error(Y_test, dummy_predictions))

print("RMSE for Dummy Regressor: ", dummy_rmse)
```

Сравнение RMSE двух моделей подтверждает, что наша модель не только адекватна, но и значительно превосходит базовую модель по качеству прогнозов.

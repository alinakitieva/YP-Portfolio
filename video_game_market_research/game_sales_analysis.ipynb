Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Цель исследования:**
1. Изучение определяющих успешность игры закономерностей на основе исторических данных о продажах игр, оценках пользователей и экспертов, жанрах и платформах.  
2. Выявление наиболее популярных жанров, платформ и возрастных категорий игр, а также их влияния на продажи. 
3. Составить портрет пользователя каждого региона
4. Разработка рекомендаций по увеличению продаж компьютерных игр.
5. Проверить следующие гипотезы: 
  - средние пользовательские рейтинги платформ Xbox One и PC одинаковые; 
  - средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

**Ход исследования:**
1. Импортирование и предварительный анализ данных, включая проверку качества данных, заполнение пропущенных значений и преобразование данных в удобный для анализа формат.

#### Удаление пропусков

````python
pd.DataFrame(round(data.isna().mean()*100,)).style.background_gradient('coolwarm')
````

Пропуски есть в следующих столбцах:
- critic_score
- user_score
- rating
- year_of_release

Причины, которые могли привести к пропускам:
- В некоторых случаях игра может не получать оценок от критиков или пользователей, потому что она малоизвестна.
- Некоторые игры могут иметь несколько версий, например, для разных платформ, и оценки могут быть доступны только для одной из версий. В этом случае значения в столбцах Critic_Score и User_Score могут быть пропущены для других версий игры.
- Нет рейтинга ESRB: Некоторые игры могут быть выпущены без рейтинга ESRB, что приводит к отсутствию значений в столбце Rating.
- Неудачный сбор данных: Пропуски могут быть результатом ошибок при сборе данных или вводе данных в таблицу.
- Ошибка при экспорте данных: В редких случаях пропуски могут появиться в результате ошибок при экспорте данных из другого источника или базы данных.

````python
data.dropna(subset=['year_of_release'], inplace=True)
````

Причины, почему заполнили пропуски определённым образом или почему не стали это делать:
- пропусков в year_of_release всего лишь 2%. Наиболее оптимальный вариант - удаление пропусков.
- пропуски в critic_Score, user_Score, rating лучше оставить как есть, так как мы не можем заполнить пропуски медианов/средним из-за того, что каждая игра уникальна и ее оценка не очень зависит от жанра, платформы и года выпуска. Также мы не можем удалить пропуски, так как их доля слишком высока.

````python
data['user_score'] = data['user_score'].replace('tbd', np.NaN)
````

Колонка `user_score` имеет тип object и в ней встречается значение "tbd", которое расшифровывается как "To Be Determined". Это значение ставится тогда, когда игра имеет недостаточное количество отзывов для подсчета рейтинга пользователей и его публикации. По своей сути это то же самое что и NaN.

#### Преобразование данных в нужные типы

````python
data['user_score'] = data['user_score'].astype('float')
data['year_of_release'] = data['year_of_release'].astype('int')
data.info()
````

Заменили тип данных в следующих столбцах:
- user_score - был тип object, заменили на float64, потому что в данных встречаются только числовые значения
- year_of_release - был тип float64, заменили на int64, потому что в данных встречаются только целые числа

#### Подсчет суммарных продаж во всех регионах и сохранение их в отдельный столбец

#### Сколько игр выпускалось в разные годы? Важны ли данные за все периоды?

````python
games_per_year = data.groupby('year_of_release').size().reset_index(name='Number_of_Games')
plt.figure(figsize=(10, 6))
plt.bar(games_per_year['year_of_release'], games_per_year['Number_of_Games'])
plt.xlabel('Year of Release')
plt.ylabel('Number of Games')
plt.title('Number of Games Released per Year')
plt.show()
````

На графике видно, что до 1995 ежегодно выпускалось очень мало игр - менее 200. С 2000 года и до 2011 можно назвать золотым периодом игровой индустрии - количество игр, выпускаемых ежегодно, росло быстрыми темпами. Пик - 2008-2009 года, после чего начался отрицательный тренд.  
Видим, что данные до 1995 можно игнорировать - игр было слишком мало.

#### Как менялись продажи по платформам? За какой характерный срок появляются новые и исчезают старые платформы?

````python
platform_sales = data.groupby('platform')['total_sales']
platform_sales.sum().sort_values().plot(kind='bar', y='total_sales', figsize=(15, 8), ec='black');
````

````python
platform_sales = data.groupby('platform')['total_sales'].sum().sort_values(ascending=False).reset_index()
top_platforms = platform_sales.head(5)['platform'].tolist()
top_platforms_data = data[data['platform'].isin(top_platforms)]
yearly_sales_by_platform = top_platforms_data.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()

plt.figure(figsize=(15, 8))
sns.lineplot(data=yearly_sales_by_platform, x='year_of_release', y='total_sales', hue='platform', marker='o', markersize=8)
plt.xlabel('Year of Release')
plt.ylabel('Total Sales')
plt.title('Total Sales by Platform per Year')
plt.legend(title='Platform')
plt.show()
````

````python
platform_sales_mean = data[data['year_of_release']>2011].groupby('platform')['total_sales'].mean().sort_values(ascending=False).reset_index()
top_platforms = platform_sales_mean.head(5)['platform'].tolist()
top_platforms_data = data[data['platform'].isin(top_platforms)]
yearly_sales_by_platform = top_platforms_data.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()

plt.figure(figsize=(15, 8))
sns.lineplot(data=yearly_sales_by_platform, x='year_of_release', y='total_sales', hue='platform', marker='o', markersize=8)
plt.xlabel('Year of Release')
plt.ylabel('Total Sales')
plt.title('Total Sales by New Platform per Year')
plt.legend(title='Platform')
plt.show()
````

**Вывод:**  
Видим, что максимальное количество продаж у консолей старого поколения - это связано с тем, что данные до 2016 года и пик игровой индустрии пришелся на 2008-2009 год.  
На графике топ 5 платформ видим, что средняя продолжительство жизни платформы - около 10 лет. 
Если посмотреть на топ 5 платформ по средним продажам с 2012 года, то видим, что лидируют в последние годы PS4 и XOne.

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```
H_0: Средние пользовательские рейтинги платформ Xbox One и PC равны.
H_a: Средние пользовательские рейтинги платформ Xbox One и PC различаются.
alpha = 0.05
```

````python
xbox_one_data = actual_data.query("platform == 'XOne'")["user_score"].dropna()
pc_data = actual_data.query("platform == 'PC'")["user_score"].dropna()

alpha = 0.05

# Проверка гипотезы о равенстве средних пользовательских рейтингов платформ Xbox One и PC
results_platforms = stats.ttest_ind(xbox_one_data, pc_data)

print(f"p-значение для гипотезы о равенстве средних пользовательских рейтингов платформ Xbox One и PC: {results_platforms.pvalue}")

if results_platforms.pvalue < alpha:
    print("Отвергаем нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC")
else:
    print("Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC")
````

#### Средние пользовательские рейтинги жанров Action и Sports разные

```
H_0: Средние пользовательские рейтинги жанров Action и Sports равны.
H_a: Средние пользовательские рейтинги жанров Action и Sports различаются.
alpha = 0.05
```

````python
action_data = actual_data.query("genre == 'Action'")["user_score"].dropna()
sports_data = actual_data.query("genre == 'Sports'")["user_score"].dropna()

# Проверка гипотезы о равенстве средних пользовательских рейтингов жанров Action и Sports
results_genres = stats.ttest_ind(action_data, sports_data)

print(f"\np-значение для гипотезы о равенстве средних пользовательских рейтингов жанров Action и Sports: {results_genres.pvalue}")

if results_genres.pvalue < alpha:
    print("Отвергаем нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports различаются")
else:
    print("Не удалось отвергнуть нулевую гипотезу: средние пользовательские рейтинги равны")
    
````

1. Гипотеза о равенстве средних пользовательских рейтингов платформ Xbox One и PC. На основе полученного p-значения, мы можем сделать вывод о том, следует ли отвергнуть нулевую гипотезу.

2. Гипотеза о разнице средних пользовательских рейтингов жанров Action и Sports. Аналогично первому случаю, на основе полученного p-значения, мы сделаем вывод о том, следует ли отвергнуть нулевую гипотезу.

Мы использовали t-критерий Стьюдента для проверки гипотез, так как это стандартный метод для сравнения средних значений двух выборок. Этот критерий предполагает нормальное распределение данных, но в данном случае мы можем считать, что выборки достаточно большие и, согласно центральной предельной теореме, они могут аппроксимироваться нормальным распределением.

**Вывод:**
1. Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC
2. Не получилось отвергнуть нулевую гипотезу о различии средних пользовательских рейтингов жанров Action и Sports
